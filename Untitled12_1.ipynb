**Part 1: Using `pyplot.plot`**

1. **Basic Line Plot:**  
   a. Generate a list of numbers from 0 to 10, incremented by 0.5.  
   b. Plot the numbers against their square values using an available style of your choice.  
   c. Label the x-axis as 'X', the y-axis as 'X^2', and give the plot a title "Plot of X vs X^2".

2. **Plot Annotations:**  
   a. Use the plot from 1.b.  
   b. Annotate the point (5, 25) with the text "This is (5,25)".  
   c. Highlight the point (5, 25) with a red circle.  

```python
import matplotlib.pyplot as plt
import numpy as np
# plt.plot()
# plt.subplots(one or more axes used to do the plot)

# 1. Basic Line Plot
# Generate a list of numbers from 0 to 10, incremented by 0.5
x = np.arange(0, 10.5, 0.5)

# Plot the numbers against their square values using an available style of your choice
plt.plot(x, x**2, linestyle='-', marker='o', color='blue')

# Label the x-axis as 'X', the y-axis as 'X^2', and give the plot a title "Plot of X vs X^2"
plt.xlabel('X')
plt.ylabel('X^2')
plt.title('Plot of X vs X^2')

# Display the plot
plt.show()


# 2. Plot Annotations, using the plot from 1.b 
plt.plot(x, x**2, linestyle='-', marker='o', color='blue')

# annotate point (5,25) with the text
plt.annotate('This is (5,25)', xy=(5, 25), fontsize=10)

# highlight point (5,25) with a red circle
plt.scatter(5, 25, s=100, c='red')

# Display the plot
plt.show()
```

Pro Tips: 

* to highlight a single point on a plot, try adding a `.scatter()` plot of a single point to your `axes`
* in general, a function that does plotting using `matplotlib` should return a `figure` (i.e. `fig`) object

```python
import matplotlib.pyplot as plt
import numpy as np

# 3. Basic Line Plot 
# generate a list of nums from -10 to 10, incremented by 1
x = np.arange(-10, 10, 1)

# create subplot object and plot the nums against their cube values
# create a figure and axes
fig, ax = plt.subplots() 

# plot the data
ax.plot(x, x**3, linestyle='-', marker='o', color='blue')

# annotate the plot
ax.set_xlabel('Y')  
ax.set_ylabel('Y^3')  
ax.set_title("Plot of Y vs Y^3")  

fig

# 4. Plot Annotations (using `ax`)
# use the plot from 3.b.  
ax.plot(x, x**3, linestyle='-', marker='o', color='orange')   
# annotate point (-5,-125) with the text
plt.annotate('This is (-5,-125)', xy=(-5, -125), fontsize=10)

# highlight point (-5,-125) with a blue circle
plt.scatter(-5, -125, s=100, c='blue')

# Display the plot
fig

'''
# 5. Combining Plots
   a. Plot both functions (X^2 and Y^3) on the same figure, using different colors for each function.  
   b. Add a legend for the two plots
   
# 6. Functional Plotting
   a. Write a function to plot its two input arguments, y and x, against one another. It should label the y-axis "activity" and the x-axis "time", and have a good looking style.  
   b. Test your function with python list and numpy 1D array inputs. (and does it work with one of each?).  
   c. Add a third argument to let the users choose between a solid line, circle markers, or both.  
   

Pro Tips: 

* to highlight a single point on a plot, try adding a `.scatter()` plot of a single point to your `axes`
* in general, a function that does plotting using `matplotlib` should return a `figure` (i.e. `fig`) object
'''
```
